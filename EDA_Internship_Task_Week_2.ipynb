```python
import pandas as pd

# Import the Excel file
data = pd.read_excel('SampleWork.xlsx', sheet_name='Sheet1', skiprows=[2])

# Extract only the first and last columns
data = data.iloc[:, [0, -1]]
print(data)
```

```python
# Set the second row as the header
header = data.iloc[0]
data = data[1:]
data.columns = header
print(data)
```

```python
# Export the modified data as a new sheet
data.to_excel('NewSheet.xlsx', index=False)
```
