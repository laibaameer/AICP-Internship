```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
df=pd.read_csv("/content/menu.csv")
```

**Q:2**

```python
df1 = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
```

```python
df1.describe()
```

```python
# Calculate the maximum value for each attribute specified
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

max_values = {}
for attribute in attributes:
    max_value = df1[attribute].max()
    max_values[attribute] = max_value

print("Maximum values for each attribute:")
for attribute, max_value in max_values.items():
    print(f"{attribute}: {max_value}")
```

**Q:3**

```python
correlation_matrix=df1.corr();
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Q:4**

```python
# Set up the figure size
plt.figure(figsize=(8, 8))

# Draw the boxplot
sns.boxplot(x='Calories', y='Category', data=df, hue='Category')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Calories')
plt.title('Boxplot of Calories by Category')



categories = df['Category'].unique()

# Create legend with category names
plt.legend(labels=categories, title='Category')

# Show the plot
plt.show()
```

**Q:5**

```python
# Dictionary to store the attribute names and corresponding items with maximum values
attributes_items_max_values = {
    'Calories': '',
    'Total Fat': '',
    'Carbohydrates': '',
    'Dietary Fiber': '',
    'Sugars': '',
    'Protein': '',
    'Vitamin A (% Daily Value)': '',
    'Vitamin C (% Daily Value)': '',
    'Calcium (% Daily Value)': '',
    'Iron (% Daily Value)': ''
}

# Iterate over each attribute and find the item with the maximum value
for attribute in attributes_items_max_values:
    max_value_index = df1[attribute].idxmax()
    max_value_item = df.loc[max_value_index, 'Item']
    max_value = df1.loc[max_value_index, attribute]
    attributes_items_max_values[attribute] = (max_value_item, max_value)

# Display the summary
for attribute, (item, value) in attributes_items_max_values.items():
    print(f"{attribute}: {item} - {value}")
```

**Q:6**

```python
# List of attributes to plot
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

# Set up the figure size
plt.figure(figsize=(16, 8))

# Iterate over each attribute and draw the stripplot
for i, attribute in enumerate(attributes, start=1):
    plt.subplot(2, 5, i)
    sns.stripplot(x='Category', y=attribute, data=df, jitter=True, hue='Category', palette='Set3')
    plt.title(attribute)
    plt.xlabel('Category')
    plt.ylabel(attribute)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.legend(loc='upper right', title='Category')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
